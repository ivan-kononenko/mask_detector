# Порівняння методів розпізнавання облич
Існує багато методів розпізнавання облич.
Ми розглянемо три з них:
- Haar cascades
- Dlib
- DNN in OpenCV

В папці `examples` знаходиться 10 зображень облич. 5 в масках, чотири без маски і один кіт.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### HAAR Cascades
Модель `haarcascade_frontalface_default.xml` можна скачати з https://github.com/opencv/opencv/tree/master/data/haarcascades.

```python
img = cv2.imread('../examples/sample9.jpeg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

face_cascade = cv2.CascadeClassifier('../cfdm_data/haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]

plt.imshow(img)
plt.show()
```

### Dlib
Dlib is a modern C++ toolkit containing machine learning algorithms and tools for creating complex software in C++ to solve real world problems. It is used in both industry and academia in a wide range of domains including robotics, embedded devices, mobile phones, and large high performance computing environments.

```python
import dlib

img = cv2.imread('../examples/sample9.jpeg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("../cfdm_data/shape_predictor_68_face_landmarks.dat")
rects = detector(gray, 1)

def shape_to_np(shape, dtype="int"):
    coords = np.zeros((shape.num_parts, 2), dtype=dtype)
    for i in range(0, shape.num_parts):
        coords[i] = (shape.part(i).x, shape.part(i).y)
    return coords


def rect_to_bb(rect):
    # take a bounding predicted by dlib and convert it
    # to the format (x, y, w, h) as we would normally do
    # with OpenCV
    x = rect.left()
    y = rect.top()
    w = rect.right() - x
    h = rect.bottom() - y

    return (x, y, w, h)


# loop over the face detections
for (i, rect) in enumerate(rects):
    # convert dlib's rectangle to a OpenCV-style bounding box
    # [i.e., (x, y, w, h)], then draw the face bounding box
    (x, y, w, h) = rect_to_bb(rect)
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
        
plt.imshow(img)
plt.show()
```

### DNN
With the release of OpenCV 3.3 the deep neural network ( dnn ) library has been substantially overhauled, allowing us to load pre-trained networks via the Caffe, TensorFlow, and Torch/PyTorch frameworks and then use them to classify input images.

Модель модуля DNN знаходиться у папці `face_detector`.

```python
net = cv2.dnn.readNet(
    "../face_detector/deploy.prototxt", 
    "../face_detector/res10_300x300_ssd_iter_140000.caffemodel"
)

image = cv2.imread("../examples/sample9.jpeg")
h, w = image.shape[:2]
blob = cv2.dnn.blobFromImage(image, 1.0, (300, 300), (104.0, 177.0, 123.0))

net.setInput(blob)

detections = net.forward()
for i in range(detections.shape[2]):
        confidence = detections[0, 0, i, 2]
        if confidence > 0.5:
            box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
            (x, y, x1, y1) = box.astype("int")
            cv2.rectangle(image, (x, y), (x1, y1), (0, 0, 255), 2)

plt.imshow(image)
```

### mtcnn
MTCNN (Multi-task Cascaded Convolutional Neural Networks) is an algorithm consisting of 3 stages, which detects the bounding boxes of faces in an image along with their 5 Point Face Landmarks (link to the paper).

```python
from mtcnn import MTCNN 
import cv2


image = cv2.cvtColor(cv2.imread("../examples/sample9.jpeg"), cv2.COLOR_BGR2RGB)
detector = MTCNN()
detections = detector.detect_faces(image)

for detection in detections:
        if detection["confidence"] > 0.5:
            cv2.rectangle(image, detection["box"], (0, 0, 255), 2)

plt.imshow(image)
```

### Make comparison
Зробимо порівняння швидсоксті та точності на зображеннях з папки `examples`

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from mtcnn import MTCNN 


face_cascade = cv2.CascadeClassifier('../cfdm_data/haarcascade_frontalface_default.xml')

net = cv2.dnn.readNet(
    "../face_detector/deploy.prototxt", 
    "../face_detector/res10_300x300_ssd_iter_140000.caffemodel"
)

detector_dlib = dlib.get_frontal_face_detector()

detector_mtcnn = MTCNN()

def detect_face_haar(image):
    faces = face_cascade.detectMultiScale(image, 1.3, 5)
    return len(faces) > 0


def detect_face_dlib(image):
   rects = detector_dlib(image, 1)
   return len(rects) == 1


def detect_face_dnn(image):
    (h, w) = image.shape[:2]
    blob = cv2.dnn.blobFromImage(image, 1.0, (300, 300), (104.0, 177.0, 123.0))
    net.setInput(blob)
    detections = net.forward()
    faces_detected = 0
    for i in range(detections.shape[2]):
        confidence = detections[0, 0, i, 2]
        if confidence > 0.5:
            faces_detected += 1        
    return faces_detected > 0


def detect_face_mtcnn(image):
    detections = detector_mtcnn.detect_faces(image)
    return len(detections) > 0
```

```python
import time


results = {
    "methods": {"haar": 0, "dlib": 0, "dnn": 0, "mtcnn": 0},
    "total": 0,
}

times = {
    "haar": [], "dlib": [], "dnn": [], "mtcnn": [],
}

iterations = 0

for filename in os.listdir("../examples/"):
    iterations += 1
    if filename.endswith(".jpeg") or filename.endswith(".jpg"):
        img = cv2.imread(os.path.join("../examples/", filename))
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        results["total"] += 1

        time_start = time.time()
        result_haar = detect_face_haar(img)
        time_end_haar = time.time()
        if result_haar is True:
            results["methods"]["haar"] += 1
            times["haar"].append(time_end_haar - time_start)
        
        
        time_start = time.time()
        result_dlib = detect_face_dlib(img)
        time_end_haar = time.time()
        if result_dlib is True:
            results["methods"]["dlib"] += 1
            time_end_dlib = time.time()
            times["dlib"].append(time_end_dlib - time_start)
        
        time_start = time.time()
        result_dnn = detect_face_dnn(img)
        time_end_haar = time.time()
        if result_dnn is True:
            results["methods"]["dnn"] += 1
            time_end_dnn = time.time()
            times["dnn"].append(time_end_dnn - time_start)
        
        time_start = time.time()
        result_mtcnn = detect_face_mtcnn(img)
        time_end_haar = time.time()
        if result_mtcnn is True:
            results["methods"]['mtcnn'] += 1
            time_end_mtcnn = time.time()
            times["mtcnn"].append(time_end_mtcnn - time_start)
```

```python
haar_acc = round(results["methods"]["haar"] / results["total"] * 100, 2)
dlib_acc = round(results["methods"]["dlib"] / results["total"] * 100, 2)
dnn_acc = round(results["methods"]["dnn"] / results["total"] * 100, 2)
mtcnn_acc = round(results["methods"]["mtcnn"] / results["total"] * 100, 2)

print(f"Haar accuracy: {haar_acc}%")
print(f"Dlib accuracy: {dlib_acc}%")
print(f"DNN accuracy: {dnn_acc}%")
print(f"MTCNN accuracy: {mtcnn_acc}%")
print("")
# print(results)



haar_time = round(sum(times["haar"]) / len(times["haar"]), 3)
dlib_time = round(sum(times["dlib"]) / len(times["dlib"]), 3)
dnn_time = round(sum(times["dnn"]) / len(times["dnn"]), 3)
mtcnn_time = round(sum(times["mtcnn"]) / len(times["mtcnn"]), 3)
print(f"Haar speed: {haar_time}s")
print(f"Dlib speed: {dlib_time}s")
print(f"DNN speed: {dnn_time}s")
print(f"MTCNN speed: {mtcnn_time}s")

# print(times)
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()

labels = results["methods"].keys()
x = results["methods"].values()
y_pos = np.arange(len(labels))

ax.barh(y_pos, x, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(labels)
ax.set_xlim(0, results["total"])
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Accuracy, #')
ax.set_title('Face detector accuracy comparison chart')

plt.show()
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()

labels = results["methods"].keys()
x = [haar_time, dlib_time, dnn_time, mtcnn_time]
y_pos = np.arange(len(labels))

ax.barh(y_pos, x, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(labels)
ax.set_xlim(0, max(x) * 1.1)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Speed, s')
ax.set_title('Face detector speed comparison chart')

plt.show()
```
